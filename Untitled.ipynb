```python
rooms = range(1, 31)
floors = range(2, 7)
names = [
    "Jones",
    "Miller",
    "Davis",
    "Garcia"
]
for name, room, floor in itertools.product(names, rooms, floors):
    number = 100 * floor + room
    print(number, name)
    if login_works(number, name):
        print("WORKS!")
        break
```
